```python
import pandas as pd

df = pd.read_csv("GDP_raw_data_file.csv", skiprows=4)
df
```

```python
years = [str(year) for year in range(1990, 2025)]
df = df.rename(columns={"Country Name": "country", "Country Code": "code"})
df = df[['country', 'code'] + years]
df['country'] = df['country'].str.strip()
df
```

```python
df_clean = df.fillna(0)
df_clean['country'] = df_clean['country'].str.strip().str.upper()
df_clean
```
